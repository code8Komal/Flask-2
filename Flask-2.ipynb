### Q2. Why is request used in Flask?

#### Ans 
 - The "request" object in Flask is used to handle and process information sent by a user's web browser when they visit a Flask-based website. It allows the website to access data like form submissions, URL parameters, and other details from the user's request.the request object helps Flask retrieve the form data so it can be processed and stored.


### Q3. Why is redirect() used in Flask?

#### Ans 
The redirect() function in Flask is used to send users to a different web page within a Flask application. It's like a signpost that tells the user's web browser to go to another URL
